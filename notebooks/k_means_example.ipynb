```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs, make_moons, make_circles

from models.k_means import KMeansClassifier
```

```python
k = 2

blobs, _ = make_blobs(n_samples=1000, n_features=2, centers=k)
moons, _ = make_moons(n_samples=1000)
noisy_circles = make_circles(n_samples=1000, factor=.5,
                                      noise=.05)
no_structure = np.random.rand(1000, 2), None
kmeans_classifier = KMeansClassifier()

for data in [blobs, moons,noisy_circles[0]]:
    kmeans_classifier.learn(data, 2)
    clusters = kmeans_classifier.infer(data)
    central_points = kmeans_classifier.central_points

    plt.rcParams['figure.figsize'] = [10, 5]
    for centroid, points in clusters.items():
        points = np.array(points)
        center = central_points[centroid]


        plt.scatter(points[:, 0], points[:, 1], marker='o')
        plt.grid()
        plt.scatter(center[0], center[1], marker='x', color="red")

    plt.show()
```
